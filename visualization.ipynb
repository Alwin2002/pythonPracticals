```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([4, 8, 1, 10])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
plt.plot(ypoints, 'o-r')
plt.show()
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
plt.plot(ypoints,ls='solid')
plt.show()
```

```python
#Use of line width
plt.plot(ypoints,lw='11')
plt.show()
```

```python
#Adding labels and titles and graph
plt.plot(ypoints)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample name')
plt.show()
```

```python
#Adding grids to graph
plt.plot(ypoints)
plt.grid(axis='x')
plt.show()
```

```python
plt.plot(ypoints)
plt.grid(axis='y',color = 'green',ls= '--',lw = 0.5)
plt.show()
```

```python
x = np.array([1,2,3])
y = np.array([5,6,7])
plt.scatter(x,y)
a = np.array([7,8,9])
b = np.array([12,13,14])
plt.scatter(a,b)
plt.show()
```

```python
#Bar graphs in matplotlib
x = np.array(['A','B','C','D'])
y = np.array([3, 8, 1, 10])
plt.bar(x,y,color='red',width=0.5)
plt.show()
```

```python
#Introduction to histograms
plt.hist(y)
plt.legend()
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.4, 0, 0, 0]
color=['c','y','k','m']
plt.pie(y, labels = mylabels,explode=myexplode,shadow='True',colors=color)
plt.legend(title='Fruits')
plt.show() 
```
